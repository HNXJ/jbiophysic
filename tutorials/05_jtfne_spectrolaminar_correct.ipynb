# JTFNE TFNE-Izhikevich spectrolaminar motif — correct

This tutorial uses the compact `jtfne` API. It is `truth_safe_unverified` and demonstrates a declared readout-level motif, not biological proof.

```
from jbiophysic import jtfne
from dataclasses import replace
```

```
cfg = jtfne.default_cfg('correct', smoke=True)
cfg = replace(cfg, vis=replace(cfg.vis, output_dir='outputs/jtfne_correct_tutorial', show=False))
model = jtfne.construct(cfg.init)
signals = jtfne.simulate(model, cfg.sim)
evaluation = jtfne.evaluate(signals, cfg.opt)
figures = jtfne.visualize(signals, cfg.vis)
evaluation.scores, evaluation.sanity
```

For full-scale runs, load/edit `configs/jtfne_spectrolaminar_correct.yaml` or `configs/jtfne_spectrolaminar_inverse.yaml` and avoid smoke defaults.
